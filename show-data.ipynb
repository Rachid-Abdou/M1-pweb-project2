```python
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df2 is your DataFrame

# Filter data for male and female
df_male = df2[df2['SEXE'] == 'H']
df_female = df2[df2['SEXE'] == 'F']

# Create a line plot for male
plt.figure(figsize=(12, 6))
sns.lineplot(x='SPECIALITE', y='MOYENNE', data=df_male, ci=None, label='Male', marker='o')

# Create a line plot for female
sns.lineplot(x='SPECIALITE', y='MOYENNE', data=df_female, ci=None, label='Female', marker='o')

plt.title('Mean MOYENNE across SPECIALITE for Male and Female')
plt.xlabel('SPECIALITE')
plt.ylabel('Mean MOYENNE')
plt.legend()
plt.show()
```

```python
# Plot 1: Bar plot for average scores
avg_scores = df.groupby('SPECIALITE')['MOYENNE'].mean()
avg_scores.plot(kind='bar', color='skyblue')
plt.title('Average Scores')
plt.xlabel('Specialty')
plt.ylabel('Average Score')
plt.show()
```

```python
# Box plot for score distribution by 'ANNEE'
plt.figure(figsize=(10, 6))
df.boxplot(column='MOYENNE', by='ANNEE', showfliers=False)
plt.title('Score Distribution by Year (ANNEE)')
plt.xlabel('Year (ANNEE)')
plt.ylabel('Score')
plt.suptitle('')  # Remove the default title added by pandas
plt.show()
```

```python
# Box plot for score distribution by 'ANNEE' and 'SPECIALITE'
plt.figure(figsize=(12, 8))
df.boxplot(column='MOYENNE', by=['ANNEE', 'SPECIALITE'], showfliers=False)
plt.title('Score Distribution by Year (ANNEE) and Specialty (SPECIALITE)')
plt.xlabel('Year (ANNEE) - Specialty (SPECIALITE)')
plt.ylabel('Score')
plt.suptitle('')  # Remove the default title added by pandas
plt.xticks(rotation=90, ha='center')  # Rotate x-axis labels for better readability
plt.show()
```

```python
# Plot 4: Histogram for distribution of scores
plt.hist(df['MOYENNE'], bins=20, color='orange', edgecolor='black')
plt.title('Score Distribution')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python

# Plot 6: Pie chart for gender distribution
gender_distribution = df['SEXE'].value_counts()
gender_distribution.plot(kind='pie', autopct='%1.1f%%', colors=['lightblue', 'lightcoral'])
plt.title('Gender Distribution')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# Plot 1: Violin plot for score distribution by gender
plt.figure(figsize=(10, 6))
sns.violinplot(x='SEXE', y='MOYENNE', data=df, palette='pastel')
plt.title('Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Score')
plt.show()
```

```python

# Plot 2: Pair plot for relationship exploration
sns.pairplot(df[['MOYENNE', 'SEXE', 'SPECIALITE']], hue='SPECIALITE', palette='Set2')
plt.suptitle('Pair Plot for Scores, Gender, and Specialty', y=1.02)
plt.show()

# Plot 3: Swarm plot for distribution of scores by gender
plt.figure(figsize=(10, 6))
sns.swarmplot(x='SEXE', y='MOYENNE', data=df, palette='dark', alpha=0.7)
plt.title('Score Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Score')
plt.show()
```
